```python
import pandas as pd
import numpy as np

file_path = r"C:\Users\Louis B\OneDrive\Bureau\ESILV\Docs_Espagne\S5\Data\Tournois_200_Duels.xlsx"
tournament = pd.read_excel(file_path)
tournament.head(60)
```

```python
names = tierList["Nom"]
names[2]
type(names)
len(names)
wChar = tournament["W Character"]
lChar = tournament["L Character"]
names[0]
```

```python
#rate des persos
nWin = np.zeros(len(names))
nPlay = np.zeros(len(names))
winRate = np.zeros(len(names))
pickRate = np.zeros(len(names))
winOverPick = np.zeros(len(names))

for i in range(len(names)):
    for j in range(len(wChar)):
        if(wChar[j] == names[i]):
            nWin[i] += 1
            nPlay[i] +=1
        elif(lChar[j] == names[i]):
            nPlay[i] += 1
    if(nPlay[i] > 2):
        winRate[i] = nWin[i] / nPlay[i]
    else :
        winRate[i] = 0
    pickRate[i] = nPlay[i] / (2 * len(wChar))
    if(pickRate[i] > 0):
        winOverPick[i] = winRate[i]/pickRate[i]

names1 = names[0:41]
names2 = names[42:83]
winRate1 = winRate[0:41]
winRate2 = winRate[42:83]
```

```python
tier = tierList["Rang"]
winRateAv = np.zeros(5)
nCharacter = np.zeros(5)
pickRateTier = np.zeros(5)
for i in range(len(names)):
    if(nPlay[i] > 2):
        if(tier[i] == "S+" or tier[i] == "S" or tier[i] == "S-"):
            winRateAv[0] += winRate[i]
            pickRateTier[0] += pickRate[i]
            nCharacter[0] += 1
        elif(tier[i] == "A+" or tier[i] == "A-"):
            winRateAv[1] += winRate[i]
            pickRateTier[1] += pickRate[i]
            nCharacter[1] += 1
        elif(tier[i] == "B+" or tier[i] == "B-"):
            winRateAv[2] += winRate[i]
            pickRateTier[2] += pickRate[i]
            nCharacter[2] += 1
        elif(tier[i] == "C+" or tier[i] == "C-"):
            winRateAv[3] += winRate[i]
            pickRateTier[3] += pickRate[i]
            nCharacter[3] += 1
        elif(tier[i] == "D+" or tier[i] == "D-"):
            winRateAv[4] += winRate[i]
            pickRateTier[4] += pickRate[i]
            nCharacter[4] += 1
        
for j in range(5):
    if(nCharacter[j] != 0):
        winRateAv[j] = winRateAv[j] / nCharacter[j]
    
tierNames = ["S","A","B","C","D"]
```

```python
pntDiff = tournament["Pnt Diff"]
avWinDiff = np.zeros(len(names))
avLoDiff = np.zeros(len(names))
avDiff = np.zeros(len(names))
nWin = np.zeros(len(names))
nLo = np.zeros(len(names))


for i in range(len(names)):
    for j in range(len(wChar)):
        if(wChar[j] == names[i]):
            avWinDiff[i] += pntDiff[j]
            nWin[i] += 1
        elif(lChar[j] == names[i]):
            avLoDiff[i] += pntDiff[j]
            nLo[i] += 1
    if(nWin[i] + nLo[i] != 0):
        avDiff[i] = (avWinDiff[i] - avLoDiff[i]) / (nWin[i] + nLo[i])
    if(nWin[i] != 0):
        avWinDiff[i] = avWinDiff[i] / nWin[i]
    if(nLo[i] != 0):
        avLoDiff[i] = avLoDiff[i] / nLo[i]
```

```python
plt.figure().add_axes([0,0,3,2]).bar(names,avWinDiff)
plt.xlabel("Character")
plt.ylabel("Average Difference of Point while winning")
plt.title("Average ease of Victory for each Character")
```

```python
wPlayers = tournament["Winner"]
lPlayers = tournament["Loser"]
```

```python
playerNbWin = np.zeros(len(players))
playerNbGame = np.zeros(len(players))
playerWinRate = np.zeros(len(players))

for i in range(len(players)):
    for j in range(len(wPlayers)):
        if(wPlayers[j] == players[i]):
            playerNbWin[i] += 1
            playerNbGame[i] += 1
        elif(lPlayers[j] == players[i]):
            playerNbGame[i] += 1
    playerWinRate[i] = playerNbWin[i] / playerNbGame[i]
```

```python
plt.figure().add_axes([0,0,3,2]).bar(players, playerWinRate)
plt.xlabel("Player")
plt.ylabel("WinRate")
plt.title("Each Player's Winrate")
```

```python
max = 0
iMax = 0
saveWinRate = 0
savePlayer = ""
for i in range(len(players)):
    max = 0
    iMax = i
    for j in range(i, len(players), 1):
        if(playerWinRate[j] > max):
            iMax = j
            max = playerWinRate[j]
    saveWinRate = playerWinRate[i]
    savePlayer = players[i]
    playerWinRate[i] = playerWinRate[iMax]
    players[i] = players[iMax]
    playerWinRate[iMax] = saveWinRate
    players[iMax] = savePlayer
```

```python
verif = []
nbPick = []
mostPlayedC = []
max = 0

for name in players:
    verif = []
    nbPick = []
    max = 0
    for j in range(len(wPlayers)):
        if(name == wPlayers[j]):
            if wChar[j] not in verif:
                verif.append(wChar[j])
                nbPick.append(0)
            nbPick[verif.index(wChar[j])] += 1
        elif(name == lPlayers[j]):
            if lChar[j] not in verif:
                verif.append(lChar[j])
                nbPick.append(0)
            nbPick[verif.index(lChar[j])] += 1
    for i in range(len(nbPick)):
        if(max < nbPick[i]):
            max = nbPick[i]
     
    #print(index(max(nbPick)))
    mostPlayedC.append(verif[nbPick.index(max)])
```

```python
len(mostPlayedC)
for i in range(79):
    print(players[i] + " : " + mostPlayedC[i])
```

```python
cTier = ''
dictionary = names.tolist()
tierValue = np.zeros(len(mostPlayedC))
for i in range(len(mostPlayedC)):
    #print(names.index(mostPlayedC[i]))
    cTier = tier[dictionary.index(mostPlayedC[i])]
    if(cTier == "S+" or cTier == "S" or cTier == "S-"):
        tierValue[i] = 0.8
    elif(cTier == "A+" or cTier == "A-"):
        tierValue[i] = 0.6
    elif(cTier == "B+" or cTier == "B-"):
        tierValue[i] = 0.4
    elif(cTier == "C+" or cTier == "C-"):
        tierValue[i] = 0.2
    elif(cTier == "D+" or cTier == "D-"):
        tierValue[i] = 0
```

```python
plt.figure().add_axes([0,0,3,2]).bar(players, tierValue)
plt.xlabel("Player")
plt.ylabel("Quality of the most played Character")
plt.title("Each Player's best Character's Quality")
```
